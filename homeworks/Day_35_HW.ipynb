```python
#合併資料
df_test[df_test.isnull().values==True]
```

```python
# 以 XXX 為例，先看 Cabin  出現值的特性
# 我們以  Age 為例，進行補植，補固定值
#補值前 平均值
print(df_test['Age'].mean()) #age平均
print(df_test['Age'].std()) #age標準差
# 連續型用分布圖行來看
g = sns.FacetGrid(df_test, col='Cabin')
g.map(sns.displot, 'Cabin', kde=False)
#補值後
df_test['Age']=df_test['Age'].fillna(0)
print(df_test['Age'].mean())
print(df_test['Age'].std()) #age標準差
df_test[df_test['Age'].isnull().values==True]
#補完就沒有遺失值了，但是看到這樣的補值效果，平均值上有很明顯的差異，
# 連續型用分布圖行來看
g = sns.FacetGrid(df_test, col='Cabin')
g.map(sns.displot, 'Age', kde=False)
```
